```python
import pandas as pd
df = pd.read_csv("data/spotify_dataset.csv", low_memory=False)
df.head()
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
df['Explicit'] = df['Explicit'].astype('category')
df['Genre'] = df['Genre'].astype('category')

df.info()
```

```python
# Estad铆sticas b谩sicas
df[['Popularity', 'Danceability', 'Energy', 'Tempo', 'Length']].describe()
```

```python
# Proporci贸n de canciones expl铆citas
df['Explicit'].value_counts(normalize=True)
```

```python
# Canciones por g茅nero (top 5)
df['Genre'].value_counts().head(5)
```

```python
import matplotlib.pyplot as plt

# Histograma de Popularidad
df['Popularity'].plot.hist(bins=50, title='Distribuci贸n de Popularidad')
plt.xlabel('Popularidad')
plt.show()
```

```python
# Tendencia de tempo promedio por a帽o
df['year'] = df['Release Date'].dt.year
df.groupby('year')['Tempo'].mean().plot(title='Tempo Promedio por A帽o')
plt.ylabel('BPM')
plt.show()
```

```python
# Conteo de canciones por emoci贸n
df['emotion'].value_counts().plot(kind='bar', title='Distribuci贸n de Emociones')
plt.ylabel('Cantidad de canciones')
plt.show()
```

```python
# Popularidad promedio por emoci贸n
df.groupby('emotion')['Popularity'].mean().sort_values(ascending=False).plot(kind='bar', title='Popularidad Promedio por Emoci贸n')
plt.ylabel('Popularidad')
plt.show()
```

```python
df.groupby('emotion')['Danceability'].mean().sort_values(ascending=False).plot(kind='bar', title='Bailabilidad por Emoci贸n')
plt.ylabel('Danceability')
plt.show()
```
